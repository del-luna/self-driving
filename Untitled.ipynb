```python
import os
import re
import cv2
import numpy as np
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
```

```python
img_path = os.listdir('lane/')
```

```python
img = cv2.imread('lane/'+img_path[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img[:,:,0], cmap= "gray")
plt.show()
```

```python
stencil = np.zeros_like(img[:,:,0])
polygon = np.array([[100,300], [260,180], [360,180], [530,300]]) # 
cv2.fillConvexPoly(stencil, polygon, 1)#   
plt.figure(figsize=(10,10))
plt.imshow(stencil, cmap= "gray")
plt.show()
```

```python
trans_img = cv2.bitwise_and(img[:,:,0], img[:,:,0], mask=stencil)#mask  
plt.figure(figsize=(10,10))
plt.imshow(trans_img, cmap= "gray")
plt.show()
```

```python
ret, thresh = cv2.threshold(trans_img, 130, 145, cv2.THRESH_BINARY) 

# plot image
plt.figure(figsize=(10,10))
plt.imshow(thresh, cmap= "gray")
plt.show()
```

```python
lines = cv2.HoughLinesP(thresh, 1, np.pi/180, 30, maxLineGap=200)
dummy = img[:,:,0].copy()

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(dummy, (x1,y1),(x2,y2), (255,0,0), 3)

plt.figure(figsize=(10,10))
plt.imshow(dummy, cmap= "gray")
plt.show()    
```
